# **Pandas Basic Programs**

```python
import pandas as pd
```

 Make list of different data types and execute it

```python
List_s =[1,2,-3,6.2," AI developer "]
print("List = ", List_s)
```

```python
# Create Series through above list by using Series()
series = pd.Series(List_s)
print("Series of List ", series)
```

```python
# Find type of Series by using type()
print ("Type of the Series is ",type(series))
```

```python
# Create Series through short method
S2= pd.Series([1,2,3,4,])
print("Series 2 =", S2)
```

```python
# Create empty Series
emp = pd.Series([])
print(emp)
```

```python
# Now change the index of S2
S2= pd.Series([1,2,3,4,], index =['a','b','c', 'd'])
print(" Series 2 = \n", S2)
```

```python
# Change datatype of S2
S2= pd.Series([1,2,3,4,], index =['a','b','c', 'd'], dtype=float)
print(" Series 2 = \n", S2)
```

```python
# Make S3 change datatype, indexes, and customize to it
S3= pd.Series([6,8,12,40,], index =['a','b','c', 'd'], dtype=float, name =" Data Values")
print(" S3 =  \n", S3)
```

```python
# create single scalar Series for single index
Scalar_s=pd.Series(0.5)
print("Single Scalar val = \n",Scalar_s)
```

```python
# create scalar Series  for making multi indexes
Scalar_s=pd.Series(0.5, index =[1,2,3,4])
print("Single Scalar val = \n",Scalar_s)
```

```python
# make Series dictionary with Odd number data values and index alphabets
dic = pd.Series({'a':1, 'b':3, 'c':5})
print("Series Dictionary = \n",dic)
```

```python
# create a ne series
S4 =pd.Series([1,'a', 3.0009])
print (S4)
```

```python
#Aaccess the value of S4 of index 0
print(S4[0])
```

```python
# Access index 2
print(S4[2])
```

```python
# Access all values by Slicing
print("Slicing = \n ", S4[0:3])
```

```python
# Find max val S4
max(S4)   # this not supporting the Str and int instances
```

```python
# make new dic Series and find its max val
dic = pd.Series([1,2,3,4,5,6])
print("Series Dictionary = \n",dic)
print ("max values of dic = ",max(dic))
```

```python
# print it values >3
print(" Dic values > 3 = \n",dic[dic>3])
```

```python
# Now add dic  and dic 2
dic + dic2
```
