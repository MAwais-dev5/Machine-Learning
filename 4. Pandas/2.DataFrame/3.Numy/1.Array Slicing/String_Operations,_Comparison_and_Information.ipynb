```python
import numpy as np
```

```python
#make 2 variables and store the strings data
Ch_name=" Welcome to M Awais Coding, "
Str1 =" where I teach and provide coding guidance."
```

```python
# Combine/ Merge both strings data by using add()
np.char.add(Ch_name,Str1)
```

```python
# to make Ch_name string in Lower case
np.char.lower(Ch_name)
```

```python
# to make Ch_name string in Upper case
np.char.upper(Ch_name)
```

```python
# to keep str1 at center with length 60
np.char.center(Str1,60)
```

```python
# Fill the  String Str1 with stars before and after
np.char.center(Str1,60,fillchar="*")
```

```python
# We Split / divide the String Ch_name into parts of list by using Split ()
np.char.split(Ch_name)
```

```python
#splits each element at newline characters by using splitlines()
np.char.splitlines("Awais\nDeveloper")
```

```python
# Make 2 new strings
str4 = "Developer AI"
str5= "Coder"
```

```python
# Joint str4 with : and str5 with /   by using join()
np.char.join([" : ", " / "], [str4, str5])
```

```python
# Replace str4 word AI into Artificial Intelligence by using replace ()
np.char.replace(str4,"AI","Artificial Intelligence")
```

```python
# Check str4 and str5 are equal or not by using equal()
np.char.equal(str4,str5)
```

```python
# take information about any string word or letter by using count()
np.char.count(str4, "e")
```
